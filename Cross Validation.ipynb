# Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=94)

model=LogisticRegression()
model.fit(x_train,y_train)
result=model.score(x_test,y_test)
print(result)

# 0.9473684210526315  at random_state = 0
# 0.9298245614035088  at random_state= 10
# 0.9824561403508771  at random_state = 94  

# So, here, we are able to understand that the accuracy will change as we change in rendom_state. So, you can't use it for categorical data. you can use it in regression model bcoz there can be answers in point(decimal) where as in categorical, answer will in different form
```

```python
for i in range(100):
    x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=94)
    model=LogisticRegression()
    model.fit(x_train,y_train)
    result=model.score(x_test,y_test)
    print(result)
```

```python
import time
start_time = time.time()

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=94)
model=LogisticRegression()
model.fit(x_train,y_train)
result=model.score(x_test,y_test)
print(result)

print("--- %s seconds ---" % (time.time() - start_time))  #<- This will show how much time machine will take for predict the answer
```

# K Fold Cross Validation

```python
# This KFold will give the range of maximum and minimum accuracy
from sklearn.model_selection import KFold,cross_val_score
```

```python
k= KFold(10)  #<- It will fold the data 10 times
```

```python
model= LogisticRegression()
result= cross_val_score(model,x,y,cv=k)
print(result)
print(np.mean(result))
```

* Here Cross_val_score will automatically find the data and aso directly find the acuraccy of our model. It will do both. But sometimes, we don't want to find acuraccy with this method. Suppose someties, we want to find the acuraccy through confusion matrix. So, at that time we only want to split the data through cross_val_score. This will be shown in next Titanic Survival Project.

```python
arr=[5,10,5,20,25,30]
final=[]
for i in arr:
    model=LogisticRegression()
    result=cross_val_score(model,x,y,cv=i)
    print(f'When k={i}',np.mean(result))
    final.append(np.mean(result))
```

```python
final
```

# Stratified K-fold Cross Validation\

It is more precise than Kfold because it takes usually unique data 

```python
from sklearn.model_selection import StratifiedKFold

skfold= StratifiedKFold(n_splits=5)
result=cross_val_score(model,x,y,cv=skfold)
print(np.mean(result))
```

# LOOCV - Leave One Out CV

```python
from sklearn.model_selection import LeaveOneOut

lc=LeaveOneOut()
result=cross_val_score(model,x,y,cv=lc)
```
